Import the dependencies:

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# transform csv to Pandas dataframe
credit_card_data = pd.read_csv('/content/credit_data.csv')
```

```python
# check first 5 rows
credit_card_data.head()
```

```python
# check last 5 rows
credit_card_data.tail()
```

```python
# dataset information
credit_card_data.info()
```

```python
 # check missing values
credit_card_data.isnull().sum()
```

```python
# check number of legit transaction and fraud transaction
credit_card_data['Class'].value_counts()
```

Too many legit data

Too less fraud data

Highly unbalanced dataset

let:

0 => legit

1 => fraud

```python
# separate legit and fraud
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
# statistical measures of legit and fraud
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
# compare legit and fraud
credit_card_data.groupby('Class').mean()
```

Under sampling

build a dataset with similar amount of legit and fraud data (492)

```python
# randomly select 492 legit data
legit_sample = legit.sample(n=492)
```

```python
# concatenating two datasets (row way)
new_dataset = pd.concat([legit_sample, fraud], axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

Splitting data into features(other data) and target(0 or 1)

```python
X = new_dataset.drop(columns= 'Class', axis=1)
Y = new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

Split data into training and testing

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Training model

```python
model = LogisticRegression()
```

```python
model.fit(X_train, Y_train)
```

Model evaluation

```python
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy = ', training_data_accuracy)
```

```python
X_test_prediction = model.predict(X_test)
testing_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy = ', testing_data_accuracy)
```

We obtained good accuracy for both train and test data
